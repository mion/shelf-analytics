```python
import math
import random
import numpy as np

import matplotlib.pyplot as plt
```

```python
def make_noisy(seq, amp=0.2, delta=None): # seq is a numpy array
    random.seed(1) # ensure some determinism
    delta = abs(seq.max() - seq.min()) if delta is None else delta
    noise_amp = amp * delta
    return np.array([(y + (random.random()*noise_amp/2) - (random.random()*noise_amp/2)) for y in seq])
```

```python
def plot(seq):
    plt.scatter([i for i in range(len(seq))], seq)
```

```python
#
# Small peak big peak
#
seq = [math.pow(i, 3) for i in range(15)] + [math.pow(15 - i, 3) for i in range(15)] + [math.pow(i, 2.5) for i in range(15)] + [math.pow(15 - i, 2.5) for i in range(15)]
seq.reverse()
small_peak_big_peak = make_noisy(np.array(seq), amp=0.35)
plot(small_peak_big_peak)
```

```python
#
# Short step
#
seq = [100*i for i in range(15)] + [100*15 for i in range(15, 30)] + [100*(45 - i) for i in range(30, 45)]
short_step = make_noisy(np.array(seq), amp=0.65)
plot(short_step)
```

```python
#
# Long step
#
seq = [100*i for i in range(10)] + [100*10 for i in range(10, 100)] + [100*(110 - i) for i in range(100, 110)]
long_step = make_noisy(np.array(seq), amp=0.45)
plot(long_step)
```
